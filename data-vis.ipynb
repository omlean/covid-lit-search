```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Raw data overview

```python
df = pd.read_csv('data/cord-19/metadata.csv', low_memory=False)
```

```python
len(df)
```

```python
df['publish_time'] = pd.to_datetime(df.publish_time)
df['year'] = df.publish_time.apply(lambda x: x.year)
```

```python
df_ = df[df.year.notnull()]
df_.year = df_.year.astype(int)
df_ = df_[df_.year >= 2000]
```

```python
year_counts = df_.groupby('year').count()
year_counts
```

```python
plt.figure(figsize=(30,15))
sns.set_style("dark")
sns.set_context("poster")
ax = sns.barplot(x=year_counts.index, y=year_counts.publish_time, palette="Blues_d")
ax.set_xticklabels(labels=year_counts.index, rotation=40)
ax.set_xlabel('')
ax.set_ylabel('')
```
